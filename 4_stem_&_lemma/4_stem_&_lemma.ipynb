```python
# spacy do not support stemming because is inaccurate, however we do need to use lemmatization
import nltk
import spacy
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python

words =["drinking", "eating" , 'ate',"smokes", 'drunk', 'comfortable','ability', 'historically', 'better' ]
for word in words:
  print(word," | ", stemming.stem(word))
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
# Using lemma
doc = nlp("drinking smokes drunk comfortable ability historically better ")
for token in doc:
  #
  print(token," | ", token.lemma_, " | ", token.lemma)
```

```python
# after using lemma some words are not chnge to their root word
doc1 = nlp("eating ate smokes comfortable ability historically better ")
for token in doc1:
  #
  print(token," | ", token.lemma_, " | ", token.lemma)
```

## Lemmatization does not include slung but this is how we can fix

```python
attribut = nlp.get_pipe('attribute_ruler')
attribut.add([[{"TEXT":"Bro"}], [{"TEXT":"Brah"}]], {"LEMMA": "Brother"})
doc4 = nlp("Bro, you gonna chill tonight? Brah yes I am! doing nothing")

for token in doc4:
 #print root word and hash
  print(token," | ", token.lemma_, " | ", token.lemma)
```

```python
# bro and Brah  become Brother after the attribute ruler
print(doc4[0].lemma_)

doc4[8].lemma_
```
