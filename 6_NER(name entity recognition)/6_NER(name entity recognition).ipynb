```python
text = "Apple investment options include stocks, bonds, mutual funds, etc .. up to 50 billion"
doc = nlp(text)
for ent in doc.ents:
  print(ent.text, "|",ent.label_, "|", spacy.explain(ent.label_))
```

```python

from spacy.tokens import Span

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Example sentence
text = "Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne in Cupertino, California on April 1, 1976."

# Process the text
doc = nlp(text)

# Suppose we want to create a span from token index 0 to 1  (Apple)
s1 = Span(doc, 0,1, label="ORG")

doc.set_ents([s1],default="unmodified")
# Print the entities and their labels
for ent in doc.ents:
    print(ent.text,"|", ent.label_)
```
