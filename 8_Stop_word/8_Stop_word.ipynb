```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
# STOP_WORDS  -- run to find all stop word
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(" I went to school early, and I went to work then made same groceries before commming home.")

# find all stop word
stop_words = [token.text for token in doc if token.is_stop]
print("Stop words in the text:", stop_words)
```

```python
 # Find and print non-stop words
non_stop_words = [token.text for token in doc if not token.is_stop]
print("Non-stop words in the text:\n", non_stop_words)
```

```python
 # Find and print non-stop words
def Preprocessing(text):
  doc= nlp(text)
  non_stop_words = [token.text for token in doc if not token.is_stop]
  return non_stop_words
```

```python
Preprocessing("Apple stock maket rise and fall in 24 hours")
```
